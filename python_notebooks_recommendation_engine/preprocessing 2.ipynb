```python
#converting release-date column to datetime
movie_metadata['release_date'] = pd.to_datetime(movie_metadata['release_date'], errors='coerce')
```

```python
# Getting only 2017 movies as we already have movies up to the year 2016 in preprocessing 1 file. 
# We don't have enough data for the movies from 2018, 2019 and 2020. 
# We'll deal with it in the upcoming preprocessing files
new_meta = movie_metadata.loc[movie_metadata.year <= 2017,['genres','id','title','year']]
```

```python
new_meta.head()
```

```python
#casting new_meta panda object to a specified datatype
new_meta['id'] = new_meta['id'].astype(int)
```

```python
data = pd.merge(new_meta, credits, on='id')
```

```python
pd.set_option('display.max_colwidth', 75)
data
```

```python
# evaluates an expression node or a conatiner display or a string containing a Python literal 
import ast
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
```

```python
#function to generate a list based on genre
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data['genres_list'] = data['genres'].map(lambda x: make_genresList(x))
```

```python
data['genres_list']
```

```python
#function to get the first actor's information
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
```

```python
#function to get the second actor's information
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
```

```python
data['actor_2_name']
```

```python
#function to get the third actor's information
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
```

```python
data['actor_3_name']
```

```python
#function to get the director's information from a list of several professions such as producers, casting-directors
def get_directors(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_directors(x))
```

```python
data['director_name']
```

```python
movie = data.loc[:,['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]
```

```python
movie
```

```python
#checking the sum of null values in every column
movie.isna().sum()
```

```python
#dropping the rows which has null values
movie = movie.dropna(how='any')
```

```python
#checking the sum of null values to verify if it still has null values
movie.isna().sum()
```

```python
#renaming the columns' names for simplicity
movie = movie.rename(columns={'genres_list':'genres'})
movie = movie.rename(columns={'title':'movie_title'})
```

```python
movie['movie_title'] = movie['movie_title'].str.lower()
```

```python
#combining all the features and adding them under one tag
movie['comb'] = movie['actor_1_name'] + ' ' + movie['actor_2_name'] + ' '+ movie['actor_3_name'] + ' '+ movie['director_name'] +' ' + movie['genres']
```
